```python
all_files_za = pd.read_csv("Classification/all_files_ZAv1.csv")
```

```python
track_and_appearance_counts = all_files_za[["main_artist", "trackAppearanceCount", "artistAppearanceCount", "End Date"]]
```

```python
track_and_appearance_counts.shape
```

```python
first_artist_appearance = track_and_appearance_counts[track_and_appearance_counts["artistAppearanceCount"] == 0]
```

```python
first_artist_appearance.shape
```

```python
first_artist_appearance_dt = first_artist_appearance[['main_artist', 'End Date']]
```

```python
first_artist_appearance_dt = first_artist_appearance_dt.drop_duplicates()
```

```python
first_artist_appearance_dt = first_artist_appearance_dt.set_index("main_artist")
```

```python
the_first_appearance_dict = first_artist_appearance_dt.to_dict(orient="index")
```

```python
the_first_appearance_dict
```

```python
# Serializing json
json_object = json.dumps(the_first_appearance_dict, indent=4)

#     Writing to sample.json
# print("za_artists/firstappearance.json")
with open("za_artists/firstappearance.json", "w") as outfile:
    outfile.write(json_object)
```
