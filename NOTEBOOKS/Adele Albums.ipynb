```python
def createalbumsjson(allartistalbums, artist, album_name, song_name, albumlyricsstr):
    albumlyricswords = albumlyricsstr.split()
    all_words = ""
    for w in albumlyricswords:
        all_words = all_words + " " + w
        
        
    allartistalbums[artist][album_name][song_name] =all_words
    
    return allartistalbums
```
